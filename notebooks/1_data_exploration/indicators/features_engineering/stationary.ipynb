Stationarity = Statistical Stability Over Time
A stationary time series has statistical properties that do not change over time:

Constant mean
Constant variance
Constant autocorrelation structure

So stationarity essentially means the series behaves the same way regardless of when you look at it.

Non-Stationary = The Rules Keep Changing
A non-stationary series has properties that drift or shift over time. This is the problem with raw asset prices — they:

Have a trending mean (prices drift upward over decades)
Have changing variance (volatility clusters — calm periods followed by turbulent ones)
Behave differently in different market regimes

This is precisely why using raw prices for backtesting is dangerous — the statistical relationships you find in one period may not hold in another.
