Historical Volatility - "what regime are we in?" 
- Models may learn different patterns in high-vol vs low-vol environments
- Def.: Standard deviation of log returns over 20 days, annualized (× √252)
- Measures how much the price swings
- High HV = risky/volatile market, low HV = calm.

Average True Range (ATR) - tells the model about price range magnitude 
- useful for the model to calibrate how significant a price move is
- Average daily price range over 14 days. 
- True Range = max of: High-Low, |High-PrevClose|, |Low-PrevClose|.
- It captures gaps (overnight jumps) that High-Low alone misses. 
- Used for stop-loss placement and position sizing.
