```python
# The ML model never sees the price levels. It only learns to predict the direction (1 = up, 0 = down) based on indicator features (RSI, MACD, etc.). This is why it's a classification problem, not regression

import numpy as np
import pandas as pd
import yfinance as yf
import sys
sys.path.insert(0, "/Users/giovanafalcao/Desktop/Projects/trading-analytics")
from utils import yfinance_fix

def fetch_price_data(ticker = TICKER, interval = INTERVAL, period = PERIOD, shift = SHIFT, lookback = LOOKBACK):
    df = yf.download(ticker, session=yfinance_fix.chrome_session, interval = interval, period = period)
    df.columns = df.columns.get_level_values(0)

    # for i in df.columns: 
    #     df[f"{i}_change"] = df[i].diff() # P(t) - P(t-1)
    #     df[f"{i}_return"] = df[i] / df[i].shift(1) - 1 # (P(t)/P(t-1)) - 1
    #     df[f"{i}_log_return"] = np.log(df[i] / df[i].shift(1)) # ln(P(t) / P(t-1)) or ln(simple_return + 1)

    # What matters for returns are only close prices, so:
    df[f"Change"] = df["Close"].diff() # P(t) - P(t-1)
    df[f"Return"] = df["Close"] / df["Close"].shift(1) - 1 # (P(t)/P(t-1)) - 1
    df[f"Log_Return"] = np.log(df["Close"] / df["Close"].shift(1)) # ln(P(t) / P(t-1)) or ln(simple_return + 1)
    df = df.reset_index(drop=True)      

    return df.iloc[-lookback:, :]

df = fetch_price_data()
df
```
